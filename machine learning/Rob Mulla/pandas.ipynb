## Subsetting Data

```python
df.columns # List columns
```

```python
# subsetting columns
# overwrite df with just the columns I want
df = df[
    [
        'title','description', 'publishTime', 
        'duration_seconds', 'viewCount', 'likeCount', 'commentCount'
        ]
]

print(df.columns)
print("\n")
df.head()
```

```python
df.shape #df has 7 columns and 247 rowa
```

```python
# Subsetting rows
df['viewCount'] > 1_000_000 # returns True of False
```

```python
# Subsetting using loc
dfSubset1= df.loc[df['viewCount'] > 1_000_000]
dfSubset1
```

```python
# Subsetting using df.query()
dfSubsetting2 = df.query('viewCount>1_000_000')
dfSubsetting2 # notice we get the same 203 rows
```

## Removing Missing Data

```python
df.query('viewCount.isna()')
```

```python
df.loc[df['viewCount'].isna()] #Keep Only missing data
```

```python
# ~ negates
# drop rows with null values
df.loc[~df['viewCount'].isna()] 
df = df.loc[~df['viewCount'].isna()] # overwrite df 
df
```

## Casting Column dtypes 

```python
df.dtypes
```

```python
# viewCount can be casted as int now that there are no more null type vales
df['viewCount']=df['viewCount'].astype('int')
df.dtypes #Notice that viewCount is now an integer 
```

```python
# Cast likeCount as integer as well
df = df.loc[~df['likeCount'].isna()] #overwrite df, drop null types
df['likeCount']=df['likeCount'].astype('int')
df.dtypes
```

```python
# Cast publishTime to datetime format
df['publishTime']=pd.to_datetime(df['publishTime'])
df.dtypes
```

```python
# Use pandas to caste str into numeric
df['likeCount']=df['likeCount'].astype('str')
df.dtypes #likeCount is now represented as object(string)
```

```python
df['likeCount']=pd.to_numeric(df['likeCount'])
df.dtypes # casted back as numeric
```
