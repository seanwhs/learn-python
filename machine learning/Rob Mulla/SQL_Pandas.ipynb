## Connect Python to Database via API
### Raw Connection to Database

![Cursor Image](cursor.png)

````python
import mysql.connector

connection = mysql.connector.connect(
    user='root',
    password='Welcome_1',
    host='localhost',
    database='employees',
    ssl_disabled=True
)

cursor = connection.cursor()

query = """  
    SELECT *    
    FROM employees e     
    JOIN salaries s     
    ON e.emp_no = s.emp_no  
    WHERE e.hire_date > '2000-01-01';
"""

cursor.execute(query)

results = []
for i, data in enumerate(cursor):
    results.append(data)

cursor.close()
connection.close()
````

## Pandas read_sql
### Require sqlalchemy and pymysql packages

````python
import sqlalchemy
from sqlalchemy import create_engine
print('sqlalchemy vresion: ', sqlalchemy.__version__)
````

````python
# Correct the connection string
con_string = 'mysql+pymysql://root:Welcome_1@localhost/employees'
engine = create_engine(con_string)
````

````python
df_read_sql = pd.read_sql(query, engine)
df_read_sql
````

## Writing to the Database

````python
max_salary = df_read_sql.groupby(
    ['first_name', 'last_name']
)['salary'].max().reset_index()

max_salary
````

````python
# store query resukts in a new table, maz_slary
max_salary.to_sql('max_salary', engine, if_exists='replace', index=False)
````

````python
max_salary_read = pd.read_sql('Select * from max_salary', engine)
max_salary_read
````

````python
max_salary_plus = max_salary.copy()
max_salary_plus['salary'] = max_salary_plus['salary'] + 1_000_000 #pay_rise: 1 million dollars
max_salary_plus 
````

````python
max_salary_plus.to_sql('max_salary', engine, if_exists='append', index=False)
max_salary_read = pd.read_sql('Select * from max_salary', engine)
max_salary_read
````
