### Question:
* What are the locations with the fastest roller-coasters?
* We should have a minimum of 10 roller-coasters in each location?

```python
df['Location']
```

```python
# Nesting filters to get our answer
df.query('Location != "Other"') \
.groupby('Location')['Speed_Mph'] \
.agg(['mean', 'count']) \
.query('count >= 10') \
.sort_values('mean')
```

```python
# Plot this as a bar chart
df.query('Location != "Other"') \
.groupby('Location')['Speed_Mph'] \
.agg(['mean', 'count']) \
.query('count >= 10') \
.sort_values('mean') \
.plot(kind="barh", title='Average Coaster Speed by Location')

plt.show()
```

```python
# Plot only mean speed
ax = df.query('Location != "Other"') \
.groupby('Location')['Speed_Mph'] \
.agg(['mean', 'count']) \
.query('count >= 10') \
.sort_values('mean')['mean'] \
.plot(kind="barh", title='Average Coaster Speed by Location')

ax.set_xlabel('Average Coaster Speed')
plt.show()
```
