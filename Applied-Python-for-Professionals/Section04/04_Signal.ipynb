```python
%matplotlib inline
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2, 4000)
signal_space = 3 * np.pi * 3 * t
```

```python
plt.plot(t, signal.sawtooth(signal_space))

plt.ylim(-2, 2)
plt.title('Sawtooth Signal')
plt.show()
```

```python
plt.plot(t, signal.square(signal_space))

plt.ylim(-2, 2)
plt.title('Square Signal')
plt.show()
```

```python

signal_space = 2 * np.pi * t
sig = np.sin(signal_space)

pwm = signal.square(30 * signal_space, duty = (sig + 1)/ 2)

```

```python
plt.subplots_adjust(wspace=0.5, hspace=0.5)

plt.subplot(2, 1, 1)
plt.plot(t, sig)
plt.ylim(-1.5, 1.5)
plt.title('Sine Wave')

plt.subplot(2, 1, 2)
plt.plot(t, pwm)
plt.ylim(-1.5, 1.5)
plt.title('PWM Wave')

plt.show()
```

```python
t = np.linspace(-1, 1, 2 * 100)

i, q, e = signal.gausspulse(t, fc=4, retquad=True, retenv=True)
plt.plot(t, i, '-')
plt.plot(t, q, '--')
plt.plot(t, e, '-.')
plt.show()
```

```python
t = np.linspace(0, 10, 501)

w = signal.chirp(t, f0=10, f1=0.1, t1= 20, method='linear')
plt.plot(t, w)
plt.title('Linear Chirp')
plt.xlabel('t (sec)')
plt.show()
```
