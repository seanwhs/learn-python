# Image Mask

```python
nrows, ncols, channels = img1.shape
row, col = np.ogrid[:nrows, :ncols]
cnt_row, cnt_col = nrows/2, ncols/2

outer_disk_mask = ((row - cnt_row) ** 2 +
                  (col - cnt_col) ** 2 > (nrows/2) ** 2)
```

```python
img2 = img1.copy()
img2[outer_disk_mask]=255
```

```python
plt.imshow(img2)
plt.axis('off')
plt.title('Masked Image')
plt.show()
```

# Histograms

```python
img1 = plt.imread('D:/Dataset/4.1.01.tiff')

r = img1[:, :, 0]
g = img1[:, :, 1]
b = img1[:, :, 2]
```

```python
plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(img1)

hist, bins = np.histogram(r.ravel(), bins= 256,range = (0, 256))
plt.subplot(2, 2, 2)
plt.title('Red Histogram')
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(g.ravel(), bins= 256,range = (0, 256))
plt.subplot(2, 2, 3)
plt.title('Green Histogram')
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(b.ravel(), bins= 256,range = (0, 256))
plt.subplot(2, 2, 4)
plt.title('Blue Histogram')
plt.bar(bins[:-1], hist)

plt.show()
```
