```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img, save_img
import numpy as np
import matplotlib.pyplot as plt
```

load_img function - loads image from a source

```python
# load image and convert to PIL object - Python Image Library
cat_img = load_img('image/cat.jpg')

#print image properties
print(cat_img.size) # 200 by 200 pixels
print(type(cat_img))
print(cat_img.mode)
print(cat_img.format)   
```

```python
#show image in viewer
cat_img.show()
```

```python
#show image inline
plt.imshow(cat_img)
plt.show()
```

img_to_array function - convert pil object to numpy function

```python
# load image and convert to PIL object - Python Image Library
cat_img = load_img('image/cat.jpg')

#convert pil image to np array
cat_img = img_to_array(cat_img)

#print image properties
print(type(cat_img))
print('---------------------------')
print(cat_img.shape) # (200,200,3) -> 200 rows, 200 columns, 3 dimensions 
print('---------------------------')
print(cat_img)   
```

```python
#show img as np array
plt.imshow(cat_img) #notice the info loss!!!
```

array_to_img function - converts np array to PIL object

```python
# convert np array to PIL object
cat_img = array_to_img(cat_img)

#show PIL image inline
plt.imshow(cat_img)
plt.show()
```

save_img function - saves np array as PIL image

```python
#load image
cat_img = load_img('image/cat.jpg', color_mode='grayscale')

#show image inline
plt.imshow(cat_img)
```

```python
#convert PIL object to np array for saving
cat_img = img_to_array(cat_img)
print(type(cat_img))
```

```python
# save image
save_img('image/cat_grey.jpg', cat_img)

#load saved image
cat_gray = load_img('image/cat_grey.jpg')

#show loaded image inline
plt.imshow(cat_gray)
```

np arrays for color and grayscale images

```python
# color image
cat_img=load_img('image/cat.jpg')
cat_img=img_to_array(cat_img)
print(cat_img.shape)

# grayscale image
cat_img=load_img('image/cat.jpg', color_mode='grayscale')
cat_img=img_to_array(cat_img)
print(cat_img.shape)
```

Change channel order of array

```python
# grayscale image
cat_img=load_img('image/cat.jpg', color_mode='grayscale')
cat_img=img_to_array(cat_img)
print(cat_img.shape)
```

```python
#change channel order of array
cat_img = np.rollaxis(cat_img, 2, 0)
print(cat_img.shape)
```
