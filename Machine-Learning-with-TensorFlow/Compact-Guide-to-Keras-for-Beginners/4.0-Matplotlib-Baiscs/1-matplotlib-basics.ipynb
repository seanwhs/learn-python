```python
import numpy as np
import matplotlib.pyplot as plt
```

Line Plot

```python
myArray = np.array([1,2,3,2,10])
plt.plot(myArray)
plt.xlabel('X-Axis Label')
plt.ylabel('y-Axis Label')
plt.title('Line Plot with MathPlotLib')
plt.show #Notice that x-axis is automatically generated
```

Bar Plot (Histogram)

```python
from matplotlib.pyplot import step
from networkx import edge_boundary


myArray = np.array([10,10,10,10,20,20,30,30,30,30,30,30,40,40,40,50,50,50])

plt.hist(myArray)
plt.xlabel('X-Label')
plt.ylabel('Y-Label: Frequency')
plt.title('Bar Plot or Histogram')
```

Scatter Plot

```python
X = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])

plt.scatter(X, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')
plt.show()
```
