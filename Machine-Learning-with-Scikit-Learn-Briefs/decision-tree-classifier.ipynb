# Decision Tree Classification on the Iris Dataset

In this notebook, we use a Decision Tree Classifier on the Iris dataset. The process includes:

1. **Importing Libraries**: We start by loading the necessary libraries for data manipulation, model training, and visualization.

2. **Preparing Data**: We load the Iris dataset, standardize the feature values to ensure consistent scaling, and split the data into training and testing sets.

3. **Training the Model**: We train a Decision Tree Classifier using the training data.

4. **Evaluating the Model**: We evaluate the model's performance by calculating its accuracy on the test set.

5. **Visualizing the Decision Tree**: We plot the decision tree to visualize how it makes decisions and save this visualization for further inspection.

```python
from sklearn import metrics
from sklearn import datasets
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = datasets.load_iris()
X = data.data
y = data.target
print(X.shape)
print(y.shape)
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.3, random_state=0)
```

```python
model = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
accuracy = metrics.accuracy_score(y_test, prediction) 
print(f'Accuracy Score:: {accuracy}')
```

```python
# Plotting the decision tree
plt.figure(figsize=(20, 10))
tree.plot_tree(
    model, 
    filled=True, 
    feature_names=data.feature_names, 
    class_names=data.target_names, 
    rounded=True, 
    fontsize=10
)
plt.title('Decision Tree Visualization')

# Save the plot to a file
plt.savefig('decision_tree.png', format='png')  # You can change the format and file name as needed

# Show the plot
plt.show()
```
