#### General Idea of Ensemble Methods
* Ensemble Learning - Combine predictions from many predictors to improve performance


## Bootstrap Sample
* Sample with replacement


#### Bagging
* Bagging - Bootstrap aggregation is often used to reduce variance of a machine learning model
* Based on Bootstrap - take B sample datasets with replcament from the training dataset (consider the training dataset as a statistical population)
* Generally used with classification/regression trees
* Individual trees are grown deep (we overfit every individual tree), so they are relatively accurate (although they have high variance)
* Get ensemble predicition by aggregating individual predictions
    * Take average for regression problems
    * Majority vote for classification problems


#### Random Forests
* Ensemble method specifically created for classification/regression trees
* Similar to bagging, each individual predictor (tree) is trained on a bootstrap sample of the training dataset
* When splitting a node from the tree, the split that is picked is the best split among a random subset of features
* This randomness makes each individual predictor slightly worst (more biased) but due to decorrelation of the predictions the overall ensemble is better

#### Boosting and AdaBoost (Adaptive Boosting)
* Based on question: Can a set of weak learners create a strong learner
* AdaBoost - popular boosting method implemented in scikit learn
* Core idea is to fit a sequence of weak learners (models) and then in every iteration re-weight the samples of the training dataset
* Eg in classification 
    * examples that are misclassified gain weight 
    * examples that are classified correctly lose weight
    * so thwe next weak learner in the sequence focuses more on the misclasified examples
