
Training And Testing Available Data

We have a dataset containing prices of used BMW cars. We are going to analyze this dataset and build a prediction function that can predict a price by taking mileage and age of the car as input. We will use sklearn train_test_split method to split training and testing dataset


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/carprices.csv')
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Mileage')
plt.ylabel('Sale Price')
plt.title('BMW Cars Sales Price')
plt.scatter(df['Mileage'], df['Sell Price($)'])
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Car Age in Years')
plt.ylabel('Sale Price')
plt.title('BMW Cars Sales Price')
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
```

```python
X = df.drop(['Sell Price($)'], axis='columns')
X
```

```python
y = df['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.285, random_state=10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
print('Prediction::')
print(prediction)
print('Actual::')
y_list = list(y_test)
print(y_list)
```

```python
model.score(X_test, y_test)
```
