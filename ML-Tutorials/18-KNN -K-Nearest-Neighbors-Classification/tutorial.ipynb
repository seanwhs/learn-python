## KNN (K Nearest Neighbors) Classification: Machine Tutorial Using Python Sklearn








```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df['flower_name'].unique()
```

```python
df0 = df[df['flower_name']=='setosa']
df1 = df[df['flower_name']=='versicolor']
df2 = df[df['flower_name']=='virginica']
```

#### Sepal length vs Sepal Width (Setosa vs Versicolor)


```python
import matplotlib.pyplot as plt

plt.title('Sepal length vs Sepal Width (Setosa vs Versicolor)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
```

#### Petal length vs Pepal Width (Setosa vs Versicolor)


```python
plt.title('Petal length vs Pepal Width (Setosa vs Versicolor)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
```

## Train test split


```python
from sklearn.model_selection import train_test_split

X = df.drop(['target','flower_name'], axis='columns')
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

#### Create KNN (K Neighrest Neighbour Classifier)


```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn.predict([[4.8,3.0,1.5,0.3]])
```

#### Plot Confusion Matrix


```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

#### Print classification report for precesion, recall and f1-score for each classes


```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
