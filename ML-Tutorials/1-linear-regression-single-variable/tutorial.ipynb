# Machine Learning With Python: Linear Regression With One Variable

##  Problem Statement: Given above data build a machine learning model that can predict home prices based on square feet area
* You can represent values as a scatter plot. 
* After that one can draw a straight line that best fits values on chart.

### 1. Build and Train Linear Regression Model

```python
import pandas as pd  # Importing pandas library for data handling
import numpy as np   # Importing numpy library for numerical operations
import matplotlib.pyplot as plt  # Importing matplotlib library for plotting
from sklearn import linear_model  # Importing linear_model module from sklearn for linear regression
```

```python
# Path to the dataset file
path_to_data = './data/homeprices.csv'

# Reading the dataset into a pandas DataFrame
df = pd.read_csv(path_to_data)

# Displaying the dataset to understand its structure
print(df.head())
```

```python
# Plotting the data points on a scatter plot
plt.xlabel("Area (sq feet)")  # Setting label for x-axis
plt.ylabel("Price (USD)")    # Setting label for y-axis
plt.title('"Home Prices" based on "Area"')  # Setting title for the plot
plt.scatter(df['area'], df['price'])    # Creating scatter plot using 'area' and 'price' columns from DataFrame
```

```python
# Creating Linear Regression Object
reg = linear_model.LinearRegression()

# Fitting the linear regression model using 'area' as feature and 'price' as target variable
reg.fit(df[['area']], df['price'])
```

```python
# Predicting the price of a home with an area of 3300 sq feet using the fitted regression model
predicted_price = reg.predict([[3300]])

# Printing the predicted price
print(predicted_price)
```

```python
# Plotting the scatter plot of 'area' against 'price'
plt.xlabel("Area (sq feet)")  # Setting label for x-axis
plt.ylabel("Price (USD)")    # Setting label for y-axis
plt.scatter(df['area'], df['price'], color='red')  # Plotting actual data points in red

# Plotting the regression line
plt.plot(df['area'], reg.predict(df[['area']]), color='blue')  # Plotting predicted values in blue

# Adding title to the plot
plt.title("Predicted Prices vs. Actual Prices")  # Setting title for the plot


```

```python
# Performing a prediction manually to verify model correctness
m = reg.coef_      # Coefficient of the linear regression model
b = reg.intercept_ # Intercept of the linear regression model
x = 3300           # Area for which we want to predict the price
y = m * x + b      # Calculating the predicted price using the linear equation y = mx + b
print(y)           # Printing the predicted price
```
