## Logistic Regression: Multiclass Classification






```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# Load the digits dataset
digits = load_digits()

# Explore the attributes of the dataset
dir(digits)
```

```python
# Display the data of the first digit
digits.data[0]
```

```python
# Display the grayscale representation of the first digit
plt.gray()
plt.matshow(digits.images[0])
```

```python
# Display grayscale representations of the first 5 digits
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
# Display the target values of the first 5 digits
digits.target[0:5]
```

## Create and train logistic regression model


```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=1)

# Print the sizes of the training and testing sets
train = len(X_train)
test = len(X_test)
print(f'Train:{train}, Test:{test}')
```

```python
# Import the logistic regression model from sklearn
from sklearn import linear_model
model = linear_model.LogisticRegression()

# Train the model using the training data
model.fit(X_train, y_train)
```

```python
# Predict the digits in the testing set
model.predict(X_test)
```

```python
# Calculate and print the accuracy score of the model
model.score(X_test, y_test)
```

```python
# Display the grayscale representation of a specific digit (index 67)
plt.matshow(digits.images[67])
```

```python
# Display the actual target value of the digit at index 67
digits.target[67]
```

```python
# Predict the digit at index 67 using the trained model
model.predict([digits.data[67]])
```

```python
# Predict digits for the first 5 data points using the trained model
model.predict(digits.data[0:5])
```

## Confusion Matrix


```python
# Predicted values of the digits in the testing set
y_predicted = model.predict(X_test)
y_predicted
```

```python
# Import pandas library
import pandas as pd

# Create a DataFrame from the predicted values
y_predicted_df = pd.DataFrame(y_predicted)

# Check the unique values in the first column of the DataFrame
unique_values = y_predicted_df[0].unique()
unique_values
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# Compute and display the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# Visualize the confusion matrix using a heatmap
import seaborn as sns
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
